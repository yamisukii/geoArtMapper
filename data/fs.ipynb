```python
import pandas as pd

df = pd.read_csv("datasetOLD.csv")
df.head()
```

```python
df = df[['nationality','e.venue','e.startdate','e.paintings','e.city','e.latitude','e.longitude']]
df.isnull().sum()
```

```python

european_countries = [
    'United Kingdom', 'Netherlands', 'Russia', 'Switzerland', 'France',
    'Italy', 'Germany', 'Czech Republic', 'Hungary', 'Ukraine', 'Romania',
    'Belgium', 'Belarus', 'Austria', 'Norway', 'Sweden', 'Finland', 'Spain',
    'Poland', 'Bulgaria', 'Georgia', 'Portugal', 'Denmark', 'Ireland',
    'Croatia', 'Slovakia', 'Greece', 'Lithuania', 'Latvia', 'Serbia',
    'Turkey', 'Slovenia', 'Estonia', 'Luxembourg', 'Montenegro', 'Armenia',
    'Bosnia and Herzegovina'
]

# Filter the DataFrame to keep only European countries
df_filtered = df[df['nationality'].isin(european_countries)]

# Drop rows where nationality is NaN
df_filtered = df_filtered.dropna(subset=['nationality'])

# Display the cleaned DataFrame
print(df_filtered)
```

```python
df_filtered.dropna(inplace=True)  # Removes columns with any NaN

df_filtered.isnull().sum()
```

```python
df_filtered.to_csv("dataset.csv", index=False)
```
